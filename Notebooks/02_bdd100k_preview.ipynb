```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# Path to training images (change path if needed)
train_dir = r"D:\ODD PROJECT\DATA\BDD\bdd100k\bdd100k\images\100k\train"

# List files and load a few
sample_images = os.listdir(train_dir)[:5]

# Plot them
plt.figure(figsize=(15, 5))
for i, img_name in enumerate(sample_images):
    img_path = os.path.join(train_dir, img_name)
    img = Image.open(img_path)
    plt.subplot(1, 5, i + 1)
    plt.imshow(img)
    plt.title(img_name)
    plt.axis("off")
plt.suptitle("Sample Images from BDD100K (train set)")
plt.show()
```
